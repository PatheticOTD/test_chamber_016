```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
```

```python
euro
```

```python
euro['year'] =  euro.data.apply( lambda x: x.year )
euro['month'] =  euro.data.apply( lambda x: x.month )
euro['day'] =  euro.data.apply( lambda x: x.day )
```

```python
euro = euro.sort_values(by = ['year', 'month', 'day'], ascending = True)[['year', 'month', 'day', 'curs']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(euro[['year', 'month', 'day']], euro.curs, test_size = 0.2, random_state = 47)
```

```python
lm.fit(euro[['year', 'month', 'day']], euro.curs)
```

```python
lm.coef_
```

```python
pred = lm.predict(x_test)
```
