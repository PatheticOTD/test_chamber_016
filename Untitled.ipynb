```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = np.arange(1,30, 0.1)
y = x * (np.random.random(290) + np.random.randint(3,10, 290))
```

```python
plt.scatter(x,y)
```

```python
df = pd.DataFrame({'x1':x, 'y':y})
df['x2'] = np.arange(290)
df['x0'] = 0
```

```python
lm = LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['x1', 'x0']], df.y, test_size = 0.2, random_state = 11)
```

```python
lm.fit(x_train, y_train)
```

```python
lm.coef_
```

```python
pred = lm.predict(x_test)
```

```python
plt.figure( figsize = (12,6))
plt.plot(x_test.x1, pred)
```
